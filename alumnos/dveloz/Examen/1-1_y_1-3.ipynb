## 1.1.1
### Primero, un algoritmo es una serie de instrucciones a seguir. 
### Segundo, una función es como una maquina a la que le metes diferentes números y dependiendo de estos, la maquina te regresara otro.
### La búsqueda lineal es una serie de instrucciones que busca los números correctos que hay que meter a una máquina para que esta te regrese el numero más chico posible. Primero, empezando de un punto en un mapa que representa a un grupo de números, buscamos en el mapa una dirección que sea adecuada para moverse y luego buscamos una distancia que determina que tanto debemos movernos en esa dirección, al llegar al nuevo punto encontramos otros números que al meterlos dentro de la maquina, esta nos devolvera un resultado más chico que el anterior. Estos pasos se repiten hasta que ya no se encuentran direcciones ni distancias que hagan más chico el resultado de la maquina (función) o si el resultado del nuevo punto es casi igual al anterior.

## 1.1.2
### El algoritmo de regiones de confianza busca hacer lo mismo que el algoritmo de búsqueda lineal pero de manera diferente. En vez de buscar una dirección en el mapa de numeros, ahora primero vamos a trazar una región en nuestra mapa donde nos podremos mover y luego escogemos la mejor dirección para reducir nuestra función, si ninguna dirección es buena, hacemos nuestra región un poquito más chica. Esto lo repetimos hasta que llegamos a los numeros que nos regresen el resultado mas chico en nuestra maquina.

### Para resolver el problema planteado definiremos una función de costo que represente la suma de las distancias entre cada punto donde haya ocurrido un crimen y la camara más cercana. De esta manera minimizamos los espacios peligrosos que no tienen camaras. Por ejemplo: si hubieran solo 8000 crimenes se colocaría una camara en cada punto donde ocurrió un crimen y nuestra función de costo sería igual a 0, teniendo de esta manera vigilancia en todos las zonas "peligrosas".

```python
import csv
import numpy as np
file = open('crime_data.csv', 'r')
file.readline()
reader = csv.reader(file)
crimes = []
for crime in reader:
    crimes.append(crime[3:5])
camara_loc  = np.random.rand(8000,2) + [19, -100]
crimen = np.array(crimes,dtype="float")

def fcosto(crime_loc, camara_loc):
    costo = 0
    contador = 1
    for crime in crime_loc:
        print("{} de 31,057".format(contador))
        distance_to_crimes = []
        for cam in camara_loc:
            dist = ((crime[0]-cam[0])**2+(crime[1]-cam[1])**2)**(1/2)
            distance_to_crimes.append(dist)
        costo = costo + min(distance_to_crimes)
        contador = contador + 1
    return costo
    
print(fcosto(crimen, camara_loc))

    

```
